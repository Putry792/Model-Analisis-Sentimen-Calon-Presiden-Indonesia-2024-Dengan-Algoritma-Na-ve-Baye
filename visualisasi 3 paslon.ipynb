```python

# Menghitung sentimen
sentimen_counts1 = paslon01['labels'].value_counts()
sentimen_counts2 = paslon02['labels'].value_counts()
sentimen_counts3 = paslon03['labels'].value_counts()

# pasl0n 01
pos_1 = paslon01[paslon01['labels'] == 'positif'].value_counts().sum()
net_1 = paslon01[paslon01['labels'] == 'netral'].value_counts().sum()
neg_1 = paslon01[paslon01['labels'] == 'negatif'].value_counts().sum()

# pasl0n 02
pos_2 = paslon02[paslon02['labels'] == 'positif'].value_counts().sum()
net_2 = paslon02[paslon02['labels'] == 'netral'].value_counts().sum()
neg_2 = paslon02[paslon02['labels'] == 'negatif'].value_counts().sum()

# pasl0n 03
pos_3 = paslon03[paslon03['labels'] == 'positif'].value_counts().sum()
net_3 = paslon03[paslon03['labels'] == 'netral'].value_counts().sum()
neg_3 = paslon03[paslon03['labels'] == 'negatif'].value_counts().sum()

labels = ['Positif', 'Netral', 'Negatif']
paslon = {
    'Anies-Muhaimin': (pos_1, net_1, neg_1),
    'Prabowo-Gibran': (pos_2, net_2, neg_2),
    'Ganjar-Mahfud': (pos_3, net_3, neg_3)
}

x = np.arange(len(labels))
width = 0.25  # the width of the bars
multiplier = 0
fig, ax = plt.subplots(layout='constrained')

for attribute, measurement in paslon.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    ax.bar_label(rects, padding=3)
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Jumlah')
ax.set_title('Sentimen Analisis Calon Presiden Indonesia 2024')
ax.set_xticks(x + width, labels)
ax.legend(loc='upper left', ncols=3)
ax.set_ylim(0, 250)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


# Hitung jumlah setiap sentimen
pos_1 = paslon01[paslon01['labels'] == 'positif'].value_counts().sum()
net_1 = paslon01[paslon01['labels'] == 'netral'].value_counts().sum()
neg_1 = paslon01[paslon01['labels'] == 'negatif'].value_counts().sum()

# pasl0n 02
pos_2 = paslon02[paslon02['labels'] == 'positif'].value_counts().sum()
net_2 = paslon02[paslon02['labels'] == 'netral'].value_counts().sum()
neg_2 = paslon02[paslon02['labels'] == 'negatif'].value_counts().sum()

# pasl0n 03
pos_3 = paslon03[paslon03['labels'] == 'positif'].value_counts().sum()
net_3 = paslon03[paslon03['labels'] == 'netral'].value_counts().sum()
neg_3 = paslon03[paslon03['labels'] == 'negatif'].value_counts().sum()

# Data untuk plot
paslon = ['Anies-Muhaimin', 'Prabowo-Gibran', 'Ganjar-Mahfud']
positif = [pos_1, pos_2, pos_3]
netral = [net_1, net_2, net_3]
negatif = [neg_1, neg_2, neg_3]


# Menentukan posisi x-axis
# Posisi x-axis
x = np.arange(len(paslon))

# Lebar bar
width = 0.25

# Membuat subplots
fig, ax = plt.subplots(figsize=(10, 6))

# Plotting the bars
bars1 = ax.bar(x - width, positif, width, label='Positif')
bars2 = ax.bar(x, netral, width, label='Netral', )
bars3 = ax.bar(x + width, negatif, width, label='Negatif')

# Menambahkan teks untuk label, judul dan label x-axis kustom
ax.set_xlabel('Calon Presiden Indonesia 2024')
ax.set_ylabel('Jumlah Sentimen')
ax.set_title('Sentimen Calon Presiden Indonesia 2024')
ax.set_xticks(x)
ax.set_xticklabels(paslon)
ax.legend()

# Menambahkan label data di atas batang
def add_labels(bars):
    for bar in bars:
        height = bar.get_height()
        ax.annotate('{}'.format(height),
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  # 3 poin offset vertikal
                    textcoords="offset points",
                    ha='center', va='bottom')

add_labels(bars1)
add_labels(bars2)
add_labels(bars3)

# Pastikan layout rapih
plt.tight_layout()
ax.grid(True, linewidth=0.5, color='gray', alpha=0.3, axis='y')
# Menampilkan plot
plt.show()
```

```python
# Hitung jumlah sentimen untuk setiap calon presiden
data_sentimen = np.array([
    [paslon01[paslon01['labels'] == 'positif'].value_counts().sum(),
     paslon01[paslon01['labels'] == 'netral'].value_counts().sum(),
     paslon01[paslon01['labels'] == 'negatif'].value_counts().sum()],
    [paslon02[paslon02['labels'] == 'positif'].value_counts().sum(),
     paslon02[paslon02['labels'] == 'netral'].value_counts().sum(),
     paslon02[paslon02['labels'] == 'negatif'].value_counts().sum()],
    [paslon03[paslon03['labels'] == 'positif'].value_counts().sum(),
     paslon03[paslon03['labels'] == 'netral'].value_counts().sum(),
     paslon03[paslon03['labels'] == 'negatif'].value_counts().sum()]
])

# Nama calon presiden
calon_presiden = ['Anies-Muhaimin', 'Prabowo-Gibran', 'Ganjar-Mahfud']

# Warna untuk setiap sentimen
warna = ['green', 'yellow', 'red']

# Lebar bar
lebar_bar = 0.35

# Membuat subplots
fig, ax = plt.subplots(figsize=(10, 6))

# Menampilkan bar chart
posisi_bar = np.arange(len(calon_presiden))  # Posisi awal bar
for i, sentimen in enumerate(data_sentimen.T):
    ax.bar(posisi_bar + i * lebar_bar, sentimen, lebar_bar, label=f"{sentimen}", color=warna[i])

# Menambahkan label sumbu
ax.set_xlabel('Calon Presiden Indonesia 2024')
ax.set_ylabel('Jumlah Sentimen')
ax.set_title('Sentimen Calon Presiden Indonesia 2024')

# Menambahkan label x-axis
ax.set_xticks(posisi_bar + lebar_bar / 2)  # Menyesuaikan posisi label
ax.set_xticklabels(calon_presiden)

# Menambahkan legenda
ax.legend()

# Menampilkan grid
ax.grid(True)

# Memastikan layout rapi
plt.tight_layout()

# Menampilkan chart
plt.show()
```
